```python
import re
import nltk
import pandas as pd
```

```python
data = pd.read_csv("D:\codes_AIR\stockerbot-export1.csv")
```

```python
data.head()
```

```python
corpus = data["text"]
```

```python
corpus.head()
```

```python
cleaned_tweets = []
for i in range(0, len(corpus)) :
    http_removal_regex = r"(?i)\b((?:https?://|www\d{0,3}[.]|[a-z0-9.\-]+[.][a-z]{2,4}/)(?:[^\s()<>]+|\(([^\s()<>]+|(\([^\s()<>]+\)))*\))+(?:\(([^\s()<>]+|(\([^\s()<>]+\)))*\)|[^\s`!()\[\]{};:'\".,<>?«»“”‘’]))"
    tweet = re.sub(http_removal_regex,' ',corpus[i])
    punctuation_removal_regex = r'[^\w\s]'
    tweet = re.sub(punctuation_removal_regex,'',tweet)
    cleaned_tweets.append(tweet.lower().split())
```

```python
cleaned_tweets[:10]
```

```python
cleaned_tweets_without_stopwords = []
stopwords = nltk.corpus.stopwords.words('english')
for tweet in cleaned_tweets:
    res = []
    for word in tweet:
        if word not in stopwords:
            res.append(word)
    cleaned_tweets_without_stopwords.append(res)
```

```python
cleaned_tweets_without_stopwords
```

```python
lemmatizer = nltk.stem.WordNetLemmatizer()

lemmatized_tweets = []
for tweet in cleaned_tweets_without_stopwords:
    res = []
    for word in tweet:
        l = lemmatizer.lemmatize(word)
        res.append(l)
    lemmatized_tweets.append(res)
```

```python
lemmatized_tweets
```

```python
stemmer = nltk.stem.PorterStemmer()

stemmed_tweets = []
for tweet in lemmatized_tweets:
    res = []
    for word in tweet:
        s = stemmer.stem(word)
        res.append(s)
    stemmed_tweets.append(res)
```

```python
stemmed_tweets
```

```python
inverted_index = dict()
for tweet in stemmed_tweets:
    for word in tweet:
        if word not in inverted_index:
            docIDs = []
            for docID, tweet in enumerate(stemmed_tweets):
                if word in tweet:
                    docIDs.append(docID)
            inverted_index[word] = docIDs
```

```python
for word in inverted_index:
    print(word, inverted_index[word])
```
